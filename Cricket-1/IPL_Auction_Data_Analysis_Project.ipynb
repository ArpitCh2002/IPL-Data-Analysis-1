## **Cleaning the Data**

In the data, the players can be segragated based on parameters -
1. Capped Players - Batters, bowlers, Allrounders and Wicket-Keepers
2. Uncapped Players - Batters, bowlers, Allrounders and Wicket-Keepers

The Uncapped players are playing for first time in the IPL, we will use their domestic performances data. For now, we will use data for only capped players for the analysis.

```python
# Segregating Batsmen Players - Capped Players

batters = data.loc[(data['Player_Type'] == "Batter")]

batters_cap = batters.loc[(batters["Capped"] == 1)]

capped_batters = batters_cap[['Player Name', 'Team', 'Nationality', 'Matches_Played', 'Runs', 'Average', 'Strike_Rate']]
```

```python
capped_batters.head(10)
```

```python
# Segregating Bowler Players - Capped Players

bowlers = data.loc[(data['Player_Type'] == "Bowler ")]

bowlers_cap = bowlers.loc[(bowlers["Capped"] == 1)]

capped_bowlers = bowlers_cap[['Player Name', 'Team', 'Nationality', 'Matches_Played', 'Wickets', 'Bowling_average', 'Economy', 'Bowling_Strike_Rate']]
```

```python
capped_bowlers.head()
```

```python
# Segregating Wiket Keepers Players - Capped Players

keepers = data.loc[(data['Player_Type'] == "Keeper")]

keepers_cap = keepers.loc[(keepers["Capped"] == 1)]

capped_keepers = keepers_cap[['Player Name', 'Team', 'Nationality', 'Matches_Played', 'Runs', 'Average', 'Strike_Rate', 'Catches', 'Run_outs', 'Stumps']]
```

```python
capped_keepers.head(10)
```

```python
# Segregating All Rounders Players - Capped Players

Allrounders = data.loc[(data['Player_Type'] == "Allrounder")]

Allrounders_cap = Allrounders.loc[(Allrounders["Capped"] == 1)]

capped_Allrounders = Allrounders_cap[['Player Name', 'Team', 'Nationality', 'Matches_Played', 'Runs', 'Average', 'Strike_Rate', 'Wickets', 'Bowling_average', 'Economy', 'Bowling_Strike_Rate']]
```

```python
capped_Allrounders.head(10)
```

```python
# Removing the null values from data

capped_batters = capped_batters.fillna(0)
capped_bowlers = capped_bowlers.fillna(0)
capped_keepers = capped_keepers.fillna(0)
capped_Allrounders = capped_Allrounders.fillna(0)
```

```python
# Checking null values from data

print(capped_batters.isna().sum())
print(capped_bowlers.isna().sum())
print(capped_keepers.isna().sum())
print(capped_Allrounders.isna().sum())
```

## **Forming Best 11 based on the data analysed**

We will consider the number of players from each category - Batters, Bowlers and All Rounders.

The combination to be used as-  3 Batter, 5 All Rounders and 3 Bowlers.

```python
# Players in the final playing XI

from IPython.display import display, HTML

batter1 = top_3_batters.loc[(top_3_batters['Player Name'] == 'David Warner ')]
batter2 = top_3_batters.loc[(top_3_batters['Player Name'] == 'Shikhar Dhawan')]
batter3 = top_3_batters.loc[(top_3_batters['Player Name'] == 'Virat Kohli')]

allrounder1 = top_3_allrounders.loc[(top_3_allrounders['Player Name'] == 'Andre Russell')]
allrounder2 = top_3_allrounders.loc[(top_3_allrounders['Player Name'] == 'Kieron Pollard ')]
allrounder3 = top_3_allrounders.loc[(top_3_allrounders['Player Name'] == 'Krunal Pandya')]
allrounder4 = top_3_allrounders.loc[(top_3_allrounders['Player Name'] == 'Dwayne Bravo ')]
allrounder5 = top_3_allrounders.loc[(top_3_allrounders['Player Name'] == 'Ravindra Jadeja ')]

bowler1 = top_3_bowlers.loc[(top_3_bowlers['Player Name'] == 'Jasprit Bumrah')]
bowler2 = top_3_bowlers.loc[(top_3_bowlers['Player Name'] == 'Yuzvendra Chahal ')]
bowler3 = top_3_bowlers.loc[(top_3_bowlers['Player Name'] == 'Bhuvneshwar Kumar')]

combined_batters = pd.concat([batter1, batter2, batter3])
display(combined_batters)

combined_allrounders = pd.concat([allrounder1, allrounder2, allrounder3, allrounder4, allrounder5])
display(combined_allrounders)

combined_bowlers = pd.concat([bowler1, bowler2, bowler3])
display(combined_bowlers)
```

```python
# Displaying the final XI players picked on the basis of given parameters

final_XI = pd.concat([combined_batters, combined_allrounders, combined_bowlers])

final_XI.fillna(0, inplace=True)

display(final_XI)
```
